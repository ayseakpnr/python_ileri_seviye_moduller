```python
# 12 aylık bir borsa verisini temsil eden grafiği oluşturun
# xticks biligisi 12 ay olacak,(ocak,şubat, vs)
# iki adet kripto para verisini gösterceksiniz
# en küçük değer 1000, en yüksek değer 50000 12 adet rastgele veri üretin 
# başlık bilgisi 2024 yılı kripto para fiyatları
#legend bilgisi ile hangi çizginin hangi para birimine ait olduğunu yazın 
# xlabel değeri "aylar"
# ylabel değeri "Fiyat/TL"



fiyat_btc = np.random.randint(1000,5000,12)
fiyat_eth = np.random.randint(1000,5000,12)

uzunluk = np.arange(fiyat_btc.shape[0])

plt.plot(fiyat_btc,label="BTC")
plt.plot(fiyat_eth,label="ETH")

plt.xlabel("aylar")
plt.ylabel("Fiyat/TL")
plt.title("2024 Yılı Kripto Para Fiyatları")


ay_bilgisi = ["Ocak","Şubat","Mart","Nisan","Mayıs","Haziran","Temmuz","Ağustos","Eylül","Ekim","Kasım","Aralık"]

plt.xticks(uzunluk,ay_bilgisi,rotation=45)
plt.legend()
plt.show()






```

```python
#bar(çubuk) grafikleri-1.gösterim:üst üste yazarak gösterme

fiyat_btc = np.random.randint(100,250,5)
fiyat_eth = np.random.randint(100,250,5)

uzunluk = np.arange(fiyat_btc.shape[0])

plt.bar(uzunluk,fiyat_btc,label="BTC")
plt.bar(uzunluk,fiyat_eth,label="ETH")
plt.xticks(uzunluk,["2020","2021","2022","2023","2024"])
plt.legend()
plt.show()
```

```python
#bar(çubuk) grafikleri-1.gösterim:yan yana yazarak gösterme

fiyat_btc = np.random.randint(100,250,5)
fiyat_eth = np.random.randint(100,250,5)

dizi_isim = ["Bitcoin","Ethereum"]

ort_btc = np.mean(fiyat_btc)
ort_eth = np.mean(fiyat_eth)

new_dizi = [ort_btc,ort_eth]

plt.bar(dizi_isim,new_dizi,color=["orange","green"],label=["BTC","ETH"])
#plt.xticks(uzunluk,["2020","2021","2022","2023","2024"])
plt.legend()

for i in range(len(dizi_isim)):
    plt.text(i,new_dizi[1],f"{new_dizi[i]:.2f}",ha="center",va="bottom")
plt.show()
```

```python
# pie chart(pasta grafiği)
#bir dizi halinde değerleri veriniz
#bu dizideki değerlerin toplamı%100olarak kabul edilir
#sonra her bir değerin yüzdesi hesaplanır ve o yüzdeye göre
#bir dilim büyüklüğü belirlenir

giderler = [180,550,350,130,290]
etiketler =["Muhasebe","AR-GE","İnsan Kaynakları","Kalite Yönetimi","Bilişim Teknolojileri(IT)"]

plt.pie(giderler,labels=etiketler)
plt.show()
```

```python
giderler = [180,550,350,130,290]
etiketler =["Muhasebe","AR-GE","İnsan Kaynakları","Kalite Yönetimi","Bilişim Teknolojileri(IT)"]

#ingilizce yüzde gösterimi için format => autopct="%1.1f%%"
#türkçe yüzde gösterimi için format => autopct="%%%.1f"
plt.pie(giderler,labels=etiketler,autopct="%%%.1f")
plt.show()
```

```python
giderler = [180,550,350,130,290]
etiketler =["Muhasebe","AR-GE","İnsan Kaynakları","Kalite Yönetimi","Bilişim Teknolojileri(IT)"]

ayirt_etme= [0,0,0.2,0,0]
#ingilizce yüzde gösterimi için format => autopct="%1.1f%%"
#türkçe yüzde gösterimi için format => autopct="%%%.1f"
plt.pie(giderler,labels=etiketler,autopct="%%%.1f",shadow=True,explode=ayirt_etme)
plt.show()
```
