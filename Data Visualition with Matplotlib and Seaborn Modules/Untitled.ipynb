```python
#bar(Ã§ubuk) grafikleri-1.gÃ¶sterim:Ã¼st Ã¼ste yazarak gÃ¶sterme

fiyat_btc = np.random.randint(100,250,5)
fiyat_eth = np.random.randint(100,250,5)

uzunluk = np.arange(fiyat_btc.shape[0])

plt.bar(uzunluk,fiyat_btc,label="BTC")
plt.bar(uzunluk,fiyat_eth,label="ETH")
plt.xticks(uzunluk,["2020","2021","2022","2023","2024"])
plt.legend()
plt.show()
```

```python
#bar(Ã§ubuk) grafikleri-1.gÃ¶sterim:yan yana yazarak gÃ¶sterme

fiyat_btc = np.random.randint(100,250,5)
fiyat_eth = np.random.randint(100,250,5)

dizi_isim = ["Bitcoin","Ethereum"]

ort_btc = np.mean(fiyat_btc)
ort_eth = np.mean(fiyat_eth)

new_dizi = [ort_btc,ort_eth]

plt.bar(dizi_isim,new_dizi,color=["orange","green"],label=["BTC","ETH"])
#plt.xticks(uzunluk,["2020","2021","2022","2023","2024"])
plt.legend()

for i in range(len(dizi_isim)):
    plt.text(i,new_dizi[1],f"{new_dizi[i]:.2f}",ha="center",va="bottom")
plt.show()
```

```python
# pie chart(pasta grafiÄi)
#bir dizi halinde deÄerleri veriniz
#bu dizideki deÄerlerin toplamÄ±%100olarak kabul edilir
#sonra her bir deÄerin yÃ¼zdesi hesaplanÄ±r ve o yÃ¼zdeye gÃ¶re
#bir dilim bÃ¼yÃ¼klÃ¼ÄÃ¼ belirlenir

giderler = [180,550,350,130,290]
etiketler =["Muhasebe","AR-GE","Ä°nsan KaynaklarÄ±","Kalite YÃ¶netimi","BiliÅim Teknolojileri(IT)"]

plt.pie(giderler,labels=etiketler)
plt.show()
```

```python
giderler = [180,550,350,130,290]
etiketler =["Muhasebe","AR-GE","Ä°nsan KaynaklarÄ±","Kalite YÃ¶netimi","BiliÅim Teknolojileri(IT)"]

#ingilizce yÃ¼zde gÃ¶sterimi iÃ§in format => autopct="%1.1f%%"
#tÃ¼rkÃ§e yÃ¼zde gÃ¶sterimi iÃ§in format => autopct="%%%.1f"
plt.pie(giderler,labels=etiketler,autopct="%%%.1f")
plt.show()
```

```python
giderler = [180,550,350,130,290]
etiketler =["Muhasebe","AR-GE","Ä°nsan KaynaklarÄ±","Kalite YÃ¶netimi","BiliÅim Teknolojileri(IT)"]

ayirt_etme= [0,0,0.2,0,0]
#ingilizce yÃ¼zde gÃ¶sterimi iÃ§in format => autopct="%1.1f%%"
#tÃ¼rkÃ§e yÃ¼zde gÃ¶sterimi iÃ§in format => autopct="%%%.1f"
plt.pie(giderler,labels=etiketler,autopct="%%%.1f",shadow=True,explode=ayirt_etme)
plt.show()
```
